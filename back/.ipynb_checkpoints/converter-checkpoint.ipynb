```python
import joblib
import xgboost as xgb
import os
```

```python
for file in [(os.path.join("./sample_xg_models/", file)) for file in os.listdir("./sample_xg_models/") if os.path.isfile(os.path.join("./sample_xg_models/", file))]:
    print(file.split("/")[-1].split(".")[0])
    s_model = joblib.load(file)
    new_model = s_model.get_booster()
    new_model.save_model(f'converted_model/{file.split("/")[-1].split(".")[0]}.bin')
```
