```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import joblib
from pprintpp import pprint as pp
```

```python
model_paths = ["DecisionTree_result","kneighbors_result","LinearRegression_result","RandomForest_result","SVR_result"]
```

```python
master_eval = {} 
for path in model_paths:
    model_data = joblib.load(f"{path}/model_data.z")
    for _,val in model_data.items():
        model_name = list(val.keys())[0]
        break
    print(model_name)
    mae_values = []
    mse_values = []
    r2_score = []
    rmse_values = []
    for i,j in model_data.items():
        mae_values.append(j[f"{model_name}"]["error_metric_all"]["mae_error"])
        mse_values.append(j[f"{model_name}"]["error_metric_all"]["mse_error"])
        r2_score.append(j[f"{model_name}"]["error_metric_all"]["r2_score"])
        rmse_values.append(j[f"{model_name}"]["error_metric_all"]["rmse_error"])

    fig = plt.figure(figsize=(10,4))
    plt.plot([x for x in range(1,16)],mae_values,"o-",label="MAE")
    plt.plot([x for x in range(1,16)],mse_values,"o-",label="MSE")
    plt.plot([x for x in range(1,16)],rmse_values,"o-",label="RMSE")
    plt.plot([x for x in range(1,16)],r2_score,"o-",label="R2")
    plt.xlabel("Number of Folds")
    plt.ylabel("Loss Value")
    plt.title(f"{model_name} Loss values")
    plt.xticks([x for x in range(1,16)])
    plt.legend()
    plt.show()
    fig.savefig(f"./plots/fold_plots/{model_name}.png")
    master_eval[model_name] = { "mae" : mae_values,
                                "mse" : mse_values,
                                "r2" : r2_score,
                                "rmse" : rmse_values}
```

```python
fig = plt.figure(figsize=(10,4))
for model,loss in master_eval.items():
    plt.plot([x for x in range(1,16)],loss["mae"],"o-",label=f"{model}")
plt.xlabel("Number of Folds")
plt.ylabel("MAE Loss Values")
plt.title(f"Loss values across models")
plt.xticks([x for x in range(1,16)])
plt.legend()
plt.show()
fig.savefig("./plots/MAE_across_models.png")
```

```python
fig = plt.figure(figsize=(10,4))
for model,loss in master_eval.items():
    plt.plot([x for x in range(1,16)],loss["mse"],"o-",label=f"{model}")
plt.xlabel("Number of Folds")
plt.ylabel("MSE Loss Values")
plt.title(f"Loss values across models")
plt.xticks([x for x in range(1,16)])
plt.legend()
plt.show()
fig.savefig("./plots/MSE_across_models.png")
```

```python
fig = plt.figure(figsize=(10,4))
for model,loss in master_eval.items():
    plt.plot([x for x in range(1,16)],loss["rmse"],"o-",label=f"{model}")
plt.xlabel("Number of Folds")
plt.ylabel("RMSE Score Values")
plt.title(f"Loss values across models")
plt.xticks([x for x in range(1,16)])
plt.legend()
plt.show()
fig.savefig("./plots/RMSE_across_models.png")
```
