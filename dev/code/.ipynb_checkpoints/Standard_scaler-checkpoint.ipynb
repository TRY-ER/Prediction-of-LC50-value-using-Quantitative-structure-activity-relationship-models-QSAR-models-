```python
import pandas as pd
# import numpy as np
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("../dataset/formatted.csv")
```

```python
tar_col = "LC50 [-LOG(mol/L)]"
temp_df = df.drop(tar_col, axis=1)
```

```python
scl = StandardScaler()
scaled_df = scl.fit_transform(temp_df)
scaled_df = pd.DataFrame(scaled_df, columns = temp_df.columns)
scaled_df
```

```python
scaled_df[tar_col] = df[tar_col]
```

```python
scaled_df.to_csv("../dataset/std_scaled_df.csv")
```
