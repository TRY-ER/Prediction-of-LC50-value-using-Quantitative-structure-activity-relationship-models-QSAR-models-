```python
import pandas as pd
import smogn

df = pd.read_csv("../dataset/formatted.csv")
tar_col = "LC50 [-LOG(mol/L)]"
smoted = smogn.smoter( 
    data = df, 
    y = tar_col,
    k = 50,
    pert = 0.05,
    rel_thres = 0.1)
# print(smoted)
smoted["isSyn"] = 1
```

```python
smoted_sample = smoted.sample(frac=0.25)
smoted_sample.reset_index(inplace=True)
smoted_sample = smoted_sample.drop("index",axis=1)
smoted_sample
```

```python
df["isSyn"] = 0
```

```python
return_df = pd.concat([df,smoted_sample], ignore_index=True)
```

```python
return_df
```

```python
return_df.to_csv("../dataset/smogn_syn_data.csv",index=False)
```
