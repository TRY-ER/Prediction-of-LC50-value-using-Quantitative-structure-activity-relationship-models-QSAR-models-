```python
import pandas as pd
# import numpy as np
from sklearn.preprocessing import StandardScaler
import joblib
```

```python
df = pd.read_csv("../dataset/smogn_syn_data.csv")
```

```python
tar_col = "LC50 [-LOG(mol/L)]"
isSyn = "isSyn"
temp_df = df.drop([tar_col, isSyn], axis=1)
```

```python
scl = StandardScaler()
scaled_df = scl.fit_transform(temp_df)
scaled_df = pd.DataFrame(scaled_df, columns = temp_df.columns)
scaled_df
```

```python
scaled_df[tar_col] = df[tar_col]
scaled_df[isSyn] = df[isSyn]
```

```python
scaled_df
```

```python
scaled_df.to_csv("../dataset/std_scaled_df.csv",index=False)
```
